```python
from sympy import *
import numpy as np
```

```python
f = lambda x : 2**x + (x**2 + 1)**(1/2) - 5
```

```python
f(1)
```

```python
f(5)
```

```python
x = symbols('x')
plot(abs(f(x).diff(x , 2)) , (x , 1 , 4))
```

```python
M = 7.5
t = symbols('t')
df = lambda x : f(t).diff().subs(t , x)
m = min(abs(df(1)) , abs(df(5)))
```

```python
x0 = 3.25
for _ in range (4) :
    x = x0 - f(x0) / df(x0)
    ss = M / 2 / m * (x - x0)**2
    x0 = x
    print(N(x , 6) , N(ss , 6))
```
