```python
for n in range(len(X) - 1) :
    h = X[n+1] - X[n]
    k1 = h * f(X[n] , y)
    k2 = h * f(X[n]+ h/2 , y + k1/2)
    k3 = h * f(X[n] + h/2 , y + k2/2)
    k4 = h * f(X[n] + h , y + k3)
    y = y + (k1 + 2*k2 + 2*k3 + k4)/6
    print(y)
```
