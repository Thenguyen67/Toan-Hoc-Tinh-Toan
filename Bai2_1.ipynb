```python
from sympy import *
import numpy as np
```

```python
f = lambda x : x**3 + sin(x) - 2
```

```python
t = symbols('t')
```

```python
df = lambda x : f(t).diff().subs(t , x)
```

```python
x0 = 2.
for _ in range (3) :
    x = x0 - f(x0) / df(x0)
    print(x)
```

```python
x = symbols('x')
```

```python
plot(abs(f(x).diff(x , 2)) , ( x , 0 , 2))
```

```python
M = 11.5
m = min(abs(df(0)) , abs(df(2)))
```

```python
x0 = 2.
for _ in range(3) :
    x = x0 - f(x0) / df(x0)
    ss = M / 2 / m * ( x - x0)**2
    x0 = x
    print(x , ss)
```

```python
x0 = 2.
n = 0
while true :
    x = x0 - f(x0) / df(x0)
    ss = M / 2 / m * ( x - x0 )**2
    x0 = x
    n += 1
    print(n , x , ss)
    if ss < 10**-6 :
        break 
```
