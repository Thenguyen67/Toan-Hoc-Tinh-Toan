```python
from sympy import *
import numpy as np
```

```python
f = lambda x , y : -y[0] + x*y[0]**Rational(1 , 2)
```

```python
y = np.array([2])
X = [2 , 2.2 , 2.4 , 2.6 , 2.8 , 3]
```

```python
for n in range(5) :
    h = X[n+1] - X[n]
    k1 = h * f(X[n] , y)
    k2 = h * f(X[n] + h/2 , y + k1/2)
    k3 = h * f(X[n] + h/2 , y + k2/2)
    k4 = h * f(X[n] + h , y + k3)
    y = y + (k1 + 2*k2 + 2*k3 + k4)/6
    print(y)
```
