```python
f = lambda x : x**3 - x**2 - 3 #khai bao ham
from sympy import * #khai bao de dung may cai diff , subs ,bla bla
```

```python
x = symbols('x') #tao bien x cho muc dich khac
plot(f(x) , ( x , 1 , 4)) #ham plot de ve do thi
```

```python
t = symbols ('t') # tao bien t de tinh , vi dao ham khong dung bien x kia duoc . Con vi sao la t thi nhin o duoi
```

```python
df = lambda x : f(t).diff().subs(t , x) # diff la dao ham , subs la thay t = x , o day tao ra df de luu dao ham cua f(x) 
#Tranh thay doi f(x) . Chon t cho de nhin
```

```python
x = 4. # Tinh f(x) voi x = 1 , x = 4 . Cai nao duong thi lay x bang cai do ( f(x) khong dao ham)
for _ in range (3) :
    x = x - f(x) / df(x)
    print(x)
```

```python
M = 22 # Nhin 22.5 vay chu lay 22 thoi nha cac mom
m = min(abs(df(1)) , abs(df(4)))
```

```python
m # dao ham cua f(t) la 3x**2 - 2*x . Khi nay ta xet f(t) tai x = 1 va x = 4 xem cai nao be hon 
```

```python
x0 = 4.
for _ in range (3) :
    x = x0 - f(x0) / df(x0)
    ss = M / 2 / m * ( x - x0)**2
    x0 = x
    print(x , ss)
```

```python
x0 = 4.
n = 0
while true :
    x = x0 - f(x0) / df(x0)
    ss = M / 2 / m * ( x - x0 )**2
    x0 = x
    n += 1
    print( n , x , ss)
    if ss < 10**-6:
        break
```
